```python
import pandas as pd

# 1. Load Dataset
df = pd.read_csv(r"C:\Users\aasth\Downloads\archive\Combined_dataset.csv")

print("Dataset Loaded Successfully!")
```

```python
# 2. Explore Data
print("\nHEAD:")
print(df.head())

print("\nTAIL:")
print(df.tail())

print("\nSHAPE:")
print(df.shape)

print("\nCOLUMNS:")
print(df.columns.tolist())

print("\nDATA TYPES:")
print(df.dtypes)

# 3. Handle Missing Values
print("\nMissing Values:")
print(df.isnull().sum())

# Fill missing values using forward fill
df.ffill(inplace=True)

# 4. Basic Operations

# Filter rows (rating > 0)
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
df = df[df['rating'] > 0]

# Select required columns
df = df[['product_id', 'title', 'initial_price',
         'final_price', 'rating', 'category']]

# 5. Remove Duplicates
print("\nRows Before Removing Duplicates:", df.shape[0])

df.drop_duplicates(inplace=True)

print("Rows After Removing Duplicates:", df.shape[0])

# Clean initial_price
df['initial_price'] = (
    df['initial_price']
    .astype(str)
    .str.replace('₹', '', regex=False)
    .str.replace(',', '', regex=False)
    .str.replace('"', '', regex=False)
    .str.strip()
)

df['initial_price'] = pd.to_numeric(
    df['initial_price'],
    errors='coerce'
)

# Clean final_price
df['final_price'] = (
    df['final_price']
    .astype(str)
    .str.replace('₹', '', regex=False)
    .str.replace(',', '', regex=False)
    .str.replace('"', '', regex=False)
    .str.strip()
)

df['final_price'] = pd.to_numeric(
    df['final_price'],
    errors='coerce'
)

# Remove rows with invalid prices
df.dropna(subset=['initial_price', 'final_price'], inplace=True)

# 6. Create Derived Column
df['quantity'] = 1
df['total_amount'] = df['final_price'] * df['quantity']

print("\nDerived Column:")
print(df[['final_price', 'quantity', 'total_amount']].head())

# 7. Save Cleaned Dataset
df.to_csv("cleaned_dataset.csv", index=False)

print("\nCleaned dataset saved successfully!")

# Final Summary
print("\nFINAL SHAPE:", df.shape)
```
